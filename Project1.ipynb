#Project Title: **Animal protection and culling**


##Purpose

Many people believe in the inherent value of animals and advocate for their protection based on ethical principles. They argue that animals have rights to be treated with compassion and respect.Protecting certain animal species can be vital for maintaining biodiversity and ecosystem balance. Conservation efforts often focus on preserving habitats and preventing the extinction of endangered species.
In certain cases, culling is deemed necessary to manage animal populations. Overpopulation can lead to ecological imbalances, depletion of resources, and conflicts with human activities. Culling aims to reduce numbers to sustainable levels.While animal protection seeks to safeguard the welfare and rights of animals, culling is a more pragmatic approach aimed at managing populations for the greater ecological or human good. Both can be contentious topics, often sparking debates about ethics, sustainability, and the balance between human needs and environmental concerns.

#**Step 5: Sorting/Searching/Validation**

```python
# Sorting Process

# Print a target column
print(data['adopt a stray dog'])
```

```python
# To sort the data in Ascending Order
data.sort_values(by='adopt a stray cat',ascending=True)
```

```python
# Define a custom function to search for rows with 'year' greater than 2000 and 'euthanize cat' less than 200
def custom_search(row):
    return row['year'] > 2000 and row['euthanize cat'] < 200

# Apply the custom function to filter rows
result = data[data.apply(custom_search, axis=1)]
print(result.head)
```

```python
# Using Group By function
# Convert 'Dt_Customer' column to datetime if it's not already in datetime format
data['euthanize dog'] = pd.to_datetime(data['euthanize dog'])

# Group the DataFrame by 'Education' and calculate aggregate statistics
education_stats = data.groupby('euthanize dog').agg({
    'euthanize cat': ['mean', 'min', 'max'],
    'year': 'sum',
    'adopt a stray cat':'sum',

})

# Display the aggregated statistics for each education level
print("Aggregate Statistics by Education Level:")
print(education_stats)
```

#**Step 6: Data Visualisation**
